## Ngày 1: Giới thiệu tổng quan về Recommendation System

### RS là gì?
- Hỗ trợ ra quyết định.
- Cung cấp giải pháp mang tính cá nhân hóa (mà không cần tìm kiếm phức tạp).

### Phân loại
- Content-based filtering: Lọc dựa theo nội dung.
- Collaborative filtering: Lọc cộng tác.
- Hybrid methods: Kết hợp cả hai phương pháp trên.

### Content-based filtering:
- Lọc dựa trên nội dung, đặc điểm và sở thích của người dùng.
- Kỹ thuật phổ biến: TF-IDF, cosine similarity, machine learning với feature embedding.

### Collaborative filtering:
- Dựa trên hành vi của người dùng mà không cần quan tâm đến nội dung, đặc điểm.
- Có hai loại chính:
    + User-based CF: Tìm người có cùng hành vi.
    + Item-based CF: Tìm các đối tượng thường được người khác quan tâm
- Kỹ thuật phổ biến: Ma trận user-item, SVD (Singular Value Decomposition), ALS (Alternating Least Squares).

### Hybrid methods:
- Kết hợp để tận dụng lợi ích của cả hai phương pháp.
- Cải thiện độ chính xác và giải quyết vấn đề cold start (người dùng mới) hoặc đối tượng mới chưa có dữ liệu.
- Các cách kết hợp:
    + Kết hợp điểm số: Tổng hợp điểm từ cả hai mô hình để chọn gợi ý.
    + Xếp tầng (Stacking): Dùng đầu ra của mô hình này làm đầu vào của mô hình kia.
    + Mô hình thống nhất (unified model): Dùng Deep learning hoặc mô hình nhúng (embedding) để học đồng thời cả hai phương pháp.

### Tóm tắt:
- Lọc theo nội dung → Dựa vào đặc điểm sản phẩm.
- Lọc cộng tác → Dựa vào hành vi người dùng.
- Mô hình kết hợp (Hybrid) → Kết hợp cả hai để tối ưu hóa đề xuất.


### Thực hành

```python
# Bộ dữ liệu MovieLens
# Bộ dữ liệu MovieLens là một bộ dữ liệu phổ biến được sử dụng trong lĩnh vực học máy và khám phá tri thức. 
# Bộ dữ liệu này chứa thông tin về các bộ phim và các đánh giá của người dùng về các bộ phim đó. 
# Bộ dữ liệu MovieLens có nhiều phiên bản khác nhau, trong đó phiên bản ml-latest-small là 
# một phiên bản nhỏ chứa 100.000 đánh giá của 600 người dùng về 9.000 bộ phim. 

# MovieLens 32M
# MovieLens 32M movie ratings. Stable benchmark dataset.
# 32 million ratings and two million tag applications applied to 87,585 movies by 200,948 users. 
# Collected 10/2023 Released 05/2024

# For more information: https://files.grouplens.org/datasets/movielens/ml-32m-README.html

import pandas as pd
import numpy as np

movies = pd.read_csv('../Datasets/MovieLens/movies.csv')
ratings = pd.read_csv('../Datasets/MovieLens/ratings.csv')

print(movies.columns)
print(ratings.columns)
```

#### Lọc theo nội dung (Content-based filtering)

```python
# Dùng TF-IDF để tính mức độ tương đồng giữa các phim dựa vào thể loại (genres)
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Xử lý dữ liệu: dùng TF-IDF để mã hóa thể loại phim
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(movies['genres'].fillna(''))

# Tính độ tương đồng giữa các phim
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Tạo DataFrame để truy vấn
movie_indices = pd.Series(movies.index, index=movies['title']).drop_duplicates()

def recommend_content_based(title, num_recommendations=5):
    idx = movie_indices[title]
    sim_scores = list(enumerate(cosine_sim([idx])))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:num_recommendations+1]
    movie_ids = [[0] for i in sim_scores]
    return movies['title'].iloc[movie_ids]

# Thử gợi ý phim dựa vào nội dung
print(recommend_content_based('Toy Story (1950)'))

# Dùng TF-IDF để mã hóa thể loại phim thành vector
# Tính độ tương đồng cosine giữa các phim
# Gợi ý phim có độ tương đồng cao với phim đầu vào
```

#### Lọc cộng tác (Collaborative filtering)

```python
# Dùng Matrix Factorization (SVD) để dự đoán phim mà người dùng có thể thích
from surprise import SVD, Dataset, Reader
from surprise.model_selection import train_test_split
from surprise import accuracy

# Chuẩn bị dữ liệu cho Surprise
reader = Reader(rating_scale=(0.5, 5.0))
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)

# Chia tập dữ liệu train-test
trainset, testset = train_test_split(data, test_size=0.2)

# Huấn luyện mô hình SVD
model = SVD()
model.fit(trainset)

# Dự đoán trên tập test
predictions = model.test(testset)

# Đánh giá mô hình
rmse = accuracy.rmse(predictions)
print(f'RMSE: {rmse}')

# Hàm dự đoán phim cho người dùng
def recommend_collaborative(user_id, num_recommendations=5):
    movie_ids = movies['movieId'].unique()
    user_movies = ratings[ratings['userId'] == user_id]['movieId']
    movie_ids = [m for m in movie_ids if m not in user_movies.values]

    predictions = [(mid, model.predict(user_id, mid).est) for mid in movie_ids]
    predictions = sorted(predictions, key=lambda x: x[1], reverse=True)[:num_recommendations]

    return movies[movies['movieId'].isin([p[0] for p in predictions])]['title']

# Gợi ý phim cho người dùng 1
print(recommend_collaborative(user_id=1))

# Dùng SVD để giảm chiều dữ liệu user-item matrix
# Huấn luyện mô hình để dự đoán rating cho các phim chưa xem
# Gợi ý phim có điểm dự đoán cao nhất
```

#### Phương pháp kết hợp (Hybrid Recommender)

```python
# Kết hợp lọc theo nội dung + lọc cộng tác bằng cách kết hợp điểm số từ cả hai mô hình
def recommend_hybrid(user_id, title, num_recommendations=5, alpha=0.5):
    # Gợi ý từ content-based
    content_recs = recommend_content_based(title, num_recommendations * 2)

    # Gợi ý từ collaborative filtering
    collab_recs = recommend_collaborative(user_id, num_recommendations * 2)

    # Kết hợp điểm số: Trung bình cộng
    recommendations = set(content_recs).union(set(collab_recs))
    scores = {movie: alpha * (title in content_recs) + (1 - alpha) * (title in collab_recs) for movie in recommendations}

    # Sắp xếp theo điểm số
    sorted_recs = sorted(scores.items(), key=lambda x: x[1], reverse=True)

    return [r[0] for r in sorted_recs[:num_recommendations]]

# Gợi ý phim cho người dùng 1 dựa trên "Toy Story (1995)"
print(recommend_hybrid(user_id=1, title='Toy Story (1995)'))

# Kết hợp cả hai phương pháp bằng cách tính điểm trung bình cộng dựa trên mức độ phù hợp
# Có thể điều chỉnh alpla để ưu tiên lọc theo nội dung hoặc lọc cộng tác

# Kết luận
# - Nếu có nội dung mô tả rõ ràng -> Dùng Content-Based Filtering
# - Nếu có nhiều dữ liệu hành vi người dùng -> Dùng Collaborative Filtering
# - Nếu muốn tối ưu hóa -> Dùng Hybrid Model
```
